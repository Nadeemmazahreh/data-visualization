# Etheream Historical Data

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("ETH_day.csv")
print(type(df['Date']))
df['Date'] = pd.to_datetime(df['Date'])
print(type(df['Date']))
df.head()
```

```python
sns.set()
fig_dims = (20, 8)
fig, ax = plt.subplots(figsize=fig_dims)
sns.lineplot(x='Date', y='Close',ax=ax, data=df)
```

# Analysis 

It is shown that Ethereum Price has surged from 390$ to 1400$ from the end of dec 2017 till mid January 2018, and then saw a sharpe decline in price.
