# COVID-19 Data 

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('covid_19_clean_complete.csv')
df['Date'] = pd.to_datetime(df['Date'])
df.groupby(by='WHO Region')
df.head()
```

```python
sns.set()
fig_dims = (20, 8)
fig, ax = plt.subplots(figsize=fig_dims)
sns.lineplot(x="Date",y="Deaths", hue="WHO Region", data=df)
```

# Analysis 

COVID-19 deaths in America had seen the most increase during the period of 2020-04 and 2020-08 compared to other regions around the world
